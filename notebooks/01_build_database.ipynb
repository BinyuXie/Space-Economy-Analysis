## Notebook 01: Build Database

**Objectives**
- Load the company database
- Inspect and clean the data
- Build an SQLite database
- Verify the database using SQL

```python
import pandas as pd
import sqlite3
```

```python
companies = pd.read_excel("../data/companies.xlsx")

companies.head()
```

```python
companies.info()
```

```python
companies.isnull().sum()
```

```python
companies.duplicated().sum()
```

```python
conn = sqlite3.connect("../data/space_economy.db")
```

```python
companies.to_sql(
    "companies",
    conn,
    if_exists="replace",
    index=False
)
```

```python
pd.read_sql(
    "SELECT * FROM companies LIMIT 5;",
    conn
)
```

```python
pd.read_sql(
    "SELECT name FROM sqlite_master WHERE type='table';",
    conn
)
```

```python
pd.read_sql(
    "PRAGMA table_info(companies);",
    conn
)
```

```python
conn.close()
```
